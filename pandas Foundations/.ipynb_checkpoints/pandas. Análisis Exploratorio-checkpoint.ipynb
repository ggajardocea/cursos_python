## Boxplots
Estos gráficos son unos de los más populares y útiles en el estudio de variables numéricas, ya que permiten hacer comparaciones rápidas entre variables y categorías de variables y entrega muchas estadísticas de resumen.
Tiene varías partes:

+ Mediana: como ya sabemos, es el valor que separa el rango de datos en dos partes de igual frecuencia
+ Primer cuartil (Q1): punto que contiene el 25% de los datos al ser ordenados
+ Tercer cuartil (Q3): punto que contiene el 75% de los datos al ser ordenados. La resta entre el Q3 y el Q1, es llamado el Rango Intercuartil (Interquartil Range o IQR) y es representado por la caja en el boxplot.
+ Bigotes: Línea que indica donde están los mínimos y máximos. Estos se calculan como (Q1 - 1.5 * IQR) y (Q3 + 1.5 * IQR) respectivamente.
+ Outliers: cualquier cosa que no esté considerada en el largo de los bigotes.

Algo a tener en consideración es que pýthon tiene varias sintaxis para hacer un boxplot, y gráficos en general, las que entregan resultados levemente diferentes, pero que es necesario tener en cuenta.

```python
# Definimos las columnas que queremos graficar
cols = ['Open', 'Volume']

# Se usa el "box" para boxplot, y subplot para graficar lado a lado las variables
aapl[cols].plot(kind = "box", subplots= True)

# Mostrar
plt.show()
```

## Histogramas, PDFs y CDFs
Un histogramas es una agrupación que permite graficar la frecuencia de intervalos definidos para una variable numérica. Esto intervalos se pueden definir de dos formas

+ Definiendo el tamaño de los intervalos (binwidth)
+ Definiendo la cantidad de intervalos (bins)

Una PDF es la función de densidad de probabilidad (Probability Density Function), se diferencia del histograma en que posee las proporciones de cada intervalo, no la frecuencia de ocurrencia. Por ello es que se interpreta como probabilidad y el histograma no.
Por otro lado, la CDF es la función de probabilidad acumulada (Cumulative Density Function). 
En términos simples, la diferencia es que la primera permite ver la probabilidad que existe de obtener cierto rango, mientras que la segunda suma la probabilidad de llegar a cierto rango.

```python
# Definir la variable a graficar 
var = ['TempHighF']

# Graficar el histograma
austin[var].plot(kind='hist', bins=30, normed = False)
plt.show()
```

```python
# Definir la variable a graficar 
var = ['TempHighF']

# Formatear el subplot de forma que la matriz de gráficos sea de dos filas, y una columna
fig, axes = plt.subplots(nrows=2, ncols=1)

# Graficar la PDF
austin[var].plot(ax=axes[0], kind='hist', bins=30, normed=True)

# Graficar la CDF
austin[var].plot(ax=axes[1], kind='hist', bins=30, normed=True, cumulative=True)
plt.show()
```

# Análisis estadístico exploratorio
Para hacer un buen análisis es necesario ser capaz de generar estadísticas de resumen, las cuales permiten en solo una métrica tener información sobre un fenómeno.
Para ellos, diversas funciones en pandas nos son útiles:

+ pd.describe(): permite obtener rangos, cuartiles, media y conteo de cada variable. Considerar que estos métodos ignoran valores perdidos. Además, cada una de las métricas puede obtenerse por si sola


```python
# Selección de columnas
stats_austin = austin[['TempHighF', 'TempLowF']]

# Trabajando con las temperaturas 
print('Función de resumen')
print(stats_austin.describe())

# Para obtener solo ciertas métricas
print('Conteo')
print(stats_austin.count())

# Mediana, no incluida en describe
print('Mediana')
print(stats_austin.median())

# Desviación estándar
print('Desviación estándar')
print(stats_austin.std())

# cuantiles
print('Cuantil 30%')
q = 0.3
print(stats_austin.quantile(q))

print('Cuantiles')
q = [0.2, 0.8]
print(stats_austin.quantile(q))

# Lo anterior queda reflejado en los boxplot
stats_austin.plot(kind = 'box')
plt.ylabel('°F')
plt.show()
```

En muchos casos resulta útil obtener métricas para cada fila, en vez de cada columna, por ejemplo, obtener el promedio entre mínimo y máximo en las Temperaturas de Austin

```python
# austin
stats_austin.head()

# Media entre máxima y mínima del dia
media = stats_austin.mean(axis = 'columns')

# head
media.head()

# Gráfico
media.plot()
```

Al trabajar con dataframes en que existen diferentes categorías, resulta útil obtener métricas y gráficos para cada una, y así tener mayor claridad en lo que implica cada una de ellas. Un buen set de datos para trabajar es el clásico 'iris', que posee tres variedades de la planta del mismo nombre: "setosa", "versicolor" y "virginica". Posee 50 ejemplos de cada una, y sus respectivas medidas de sépalos y pétalos.
El dataset está dentro de la libreria sklearn, muy utilizada para aprendizaje automático, y que se verá próximamente en el curso.

```python
# Se carga la librería que contiene el dataset y se hacen algunas transformaciones para pasarlo a dataframe
from sklearn.datasets import load_iris
iris = load_iris()

# Se convierte a dataframe, no pescar mucho, fuera del scope de la clase
df= pd.DataFrame(data = np.c_[iris['data'], iris['target']],
                 columns= iris['feature_names'] + ['target'])

df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df = df.drop(columns = 'target')
df.head()
```

Es posible hacer varias operaciones con filtros, por ejemplo

+ ¿Cual es el promedio de ancho de pétalo de virginica?
+ ¿Cuantos registros hay de setosa?
+ ¿Cual es mayor, el promedio de largo de sépalo de setosa o versicolor?

```python
# Promedio de ancho de pétalo de virginica
print(df[df['species'] == 'virginica']['petal width (cm)'].mean())

# Registros hay de setosa
print(df[df['species'] == 'setosa'].count())

# ¿Cual es mayor, el promedio de largo de sépalo de setosa o versicolor?
pls_setosa = df[df['species'] == 'setosa']['sepal length (cm)'].mean()
pls_versicolor = df[df['species'] == 'versicolor']['sepal length (cm)'].mean()
print('Promedio de largo de sépalo en setosa: ' + str(pls_setosa))
print('Promedio de largo de sépalo en versicolor: ' + str(pls_versicolor))
```

Una forma de obtener una intuición rápida de lo que implica cada categoría, para cada una de las columnas, es comparar con el promedio de la población. Por ejemplo para setosa, podemos observar que en clase setosa el sépalo tiende a ser más corto y ancho, y el pétalo más corto y delgado.

```python
# Calcular la media global para cada característica
media_global = df.mean()

# Obtener el dataset solo de la clase setosa
setosa = df[df['species'] == 'setosa']

# Calcular la media de las características de la clase setosa
media_setosa = setosa.mean()

# Observar las diferencias
print(media_setosa - media_global)
```

Análisis como el anterior suelen ser mejor expresados en gráficos, por ejemplo para los largos del pétalo, es posible obtener un gráfico para cada clase.
pd: también es posible hacerlo de una como se observó en los boxplots de gapminder, pero eso implica pivoteos con la data que veremos en 

```python
# Se separan el espacio en que se graficará en la cantidad de gráficos que se quieren, en este caso 3. 
# Se especifica además que el eje y tenga las mismas medidas con sharey = True
fig, axes = plt.subplots(nrows=1, ncols=3, sharey=True)

# Se genera el gráfico de setosa, en el argumento ax, se pone que eje de los anteriormente definidos se quiere utilizar
df.loc[df['species'] == 'setosa'].plot(ax=axes[0], y='sepal length (cm)', kind='box')

# Se genera el gráfico de virginica
df.loc[df['species'] == 'virginica'].plot(ax=axes[1], y='sepal length (cm)', kind='box')

# Se genera el gráfico de versicolor
df.loc[df['species'] == 'versicolor'].plot(ax=axes[2], y='sepal length (cm)', kind='box')

# Mostrar el gráfico
plt.show()
```
